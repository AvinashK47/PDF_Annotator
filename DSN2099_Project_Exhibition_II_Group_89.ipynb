# Preprocessing Data

We take the raw data in `data/raw_data` and run it through text preprocessing. This returns us a processed dataset that breaks down the data on a sentence-by-sentence basis.

Each entry contains the text, its classification, previous and next sentences (where applicable), and additional metadata (not necessary for training).

The processed dataset is saved in `data/processed_data`.

## Utils

```python
tokenizer = nltk.data.load('tokenizers/punkt/english.pickle')
# common legal abbreviations
legal_abbrevs = {'no', 'vs', 'v', 'et', 'al', 'etc', 'i.e', 'e.g', 'art', 'arts',
                 'para', 'paras', 'vol', 'vols', 'fig', 'figs', 'ref', 'refs',
                 'sec', 'secs', 'mr', 'mrs', 'dr', 'hon', 'jr', 'sr'}
tokenizer._params.abbrev_types.update(legal_abbrevs)

LEGAL_CITATIONS = r'\b\d+\s+[A-Za-z\.]+\s+\d+\b|\b\[\d{4}\]\s*[A-Z]+\s*\d+|\b\(\d{4}\)\s*\d+\s*[A-Z]+\s*\d+'
BOILERPLATE_STARTS = {
    'in the case of', 'it is noted that', 'it may be noted', 'it is observed',
    'it is pertinent to note', 'it is well settled', 'it is well known'
}

def normalize_text(text: str) -> str:
    # standardizing quotes and dashes
    text = re.sub(r'[''`]', "'", text)
    text = re.sub(r'["""]', '"', text)
    text = re.sub(r'[‐‑‒–—―]', '-', text)

    # removing multiple spaces
    text = re.sub(r'\s+', ' ', text)

    # normalizing citations
    text = re.sub(LEGAL_CITATIONS, '[CITATION]', text)

    return text.strip()

def is_boilerplate(sentence: str) -> bool:
    lower_sent = sentence.lower()
    return any(lower_sent.startswith(pattern) for pattern in BOILERPLATE_STARTS)

def extract_sections(case: dict) -> dict:
    def clean_text(value) -> str:
        if value is None:
            return ""
        return str(value).strip()

    # combine similar fields
    sections = {
        "id": clean_text(case.get("Document ID", "")),
        "case_name": clean_text(case.get("Case Name", "")),
        "facts": " ".join(filter(None, [
            clean_text(case.get("Facts", "")),
            clean_text(case.get("Facts ", "")),  # in case of extra space
            clean_text(case.get("Background", ""))
        ])),
        "issues": " ".join(filter(None, [
            clean_text(case.get("Issues", "")),
            clean_text(case.get("Questions of Law", ""))
        ])),
        "precedent_analysis": " ".join(filter(None, [
            clean_text(case.get("Precedent Analysis", "")),
            clean_text(case.get("Prior Cases", "")),
            clean_text(case.get("Case Law", ""))
        ])),
        "court_reasoning": " ".join(filter(None, [
            clean_text(case.get("Court's Reasoning", "")),
            clean_text(case.get("Analysis", "")),
            clean_text(case.get("Legal Analysis", ""))
        ])),
        "conclusion": " ".join(filter(None, [
            clean_text(case.get("Conclusion", "")),
            clean_text(case.get("Judgment", "")),
            clean_text(case.get("Final Decision", ""))
        ]))
    }

    return {k: " ".join(v.split()) if isinstance(v, str) else str(v) for k, v in sections.items()}

def clean_serial_numbers(text: str) -> str:
    # temporarily replacing periods to preserve abbrevaitions
    abbrev_patterns = r'(?:No\.|vs\.|v\.|et al\.|etc\.|i\.e\.|e\.g\.|Art\.|Vol\.|Fig\.|Ref\.|Sec\.|Mr\.|Mrs\.|Dr\.|Hon\.|Jr\.|Sr\.)'
    text = re.sub(abbrev_patterns, lambda m: m.group().replace('.', '@'), text, flags=re.IGNORECASE)

    # removing roman numerals numbering
    text = re.sub(r'^\s*\d+\.|\(\w+\)|\[\w+\]|\d+\)', '', text)
    text = re.sub(r'^\s*[IVXLCDM]+\.', '', text)
    text = re.sub(r'^\s*\([iv]+\)', '', text)
    text = re.sub(r'^\s*[A-Za-z]\.\s+', '', text)  # single letter points

    # put periods back
    text = text.replace('@', '.')
    return text.strip()

def count_words(text: str) -> int:
    # we are counting common legal abbreviations as single words
    abbrev_patterns = r'(?:No\.|vs\.|v\.|et al\.|etc\.|i\.e\.|e\.g\.|Art\.|Vol\.|Fig\.|Ref\.|Sec\.|Mr\.|Mrs\.|Dr\.|Hon\.|Jr\.|Sr\.)'
    text = re.sub(abbrev_patterns, lambda m: m.group().replace('.', '@'), text, flags=re.IGNORECASE)
    return len(text.split())

def is_likely_acronym_part(text: str) -> bool:
    return bool(re.match(r'^[A-Z]$|^[A-Z]{2,3}$', text.strip()))
```

## Without SpaCy

```python
def split_into_sentences(text: str) -> list[str]:
    if not text.strip():
        return []

    lines = text.split('\n')
    sentences = []

    MIN_WORDS = 6

    for line in lines:
        if line.strip():
            cleaned_line = clean_serial_numbers(line)
            if cleaned_line:
                raw_sentences = tokenizer.tokenize(cleaned_line)

                merged_sentences = []
                current_sentence = ""

                for s in raw_sentences:
                    if is_likely_acronym_part(s):
                        current_sentence = (current_sentence + " " + s).strip()
                    else:
                        if current_sentence:
                            if count_words(current_sentence) >= MIN_WORDS:
                                merged_sentences.append(current_sentence)
                            current_sentence = s
                        else:
                            current_sentence = s

                if current_sentence and count_words(current_sentence) >= MIN_WORDS:
                    merged_sentences.append(current_sentence)

                sentences.extend(merged_sentences)

    return sentences

def sentence_splitter(text: str) -> list[str]:
    # wrapper function for annotation
    return split_into_sentences(text)
```

```python
def get_sentence_context(sentences: list[str], current_idx: int, window: int = 1) -> dict:
    start_idx = max(0, current_idx - window)
    end_idx = min(len(sentences), current_idx + window + 1)

    return {
        "previous": sentences[start_idx:current_idx],
        "next": sentences[current_idx + 1:end_idx]
    }

def process_case_sections(case: dict) -> list[dict]:
    sections = extract_sections(case)
    samples = []

    section_types = {
        "facts": "FACT",
        "issues": "ISSUE",
        "precedent_analysis": "PRECEDENT",
        "court_reasoning": "REASONING",
        "conclusion": "CONCLUSION"
    }

    for section_name, label in section_types.items():
        text = sections[section_name]
        text = normalize_text(text)
        sentences = split_into_sentences(text)

        for idx, sentence in enumerate(sentences):
            if sentence.strip() and not is_boilerplate(sentence):
                context = get_sentence_context(sentences, idx)
                sample = {
                    "text": sentence.strip(),
                    "label": label,
                    "metadata": {
                        "case_id": sections["id"],
                        "case_name": sections["case_name"],
                        "section": section_name,
                        "context": context,
                        "position": f"{idx + 1}/{len(sentences)}",
                        "has_citation": bool(re.search(LEGAL_CITATIONS, sentence))
                    }
                }
                samples.append(sample)

    return samples

def balance_samples(samples, max_per_category=None):
    label_groups = {}
    for sample in samples:
        label = sample['label']
        if label not in label_groups:
            label_groups[label] = []
        label_groups[label].append(sample)

    if max_per_category is None:
        max_per_category = min(len(group) for group in label_groups.values())

    balanced_samples = []
    for label, group in label_groups.items():
        balanced_samples.extend(random.sample(group, max_per_category))

    random.shuffle(balanced_samples)
    return balanced_samples

def preprocess_sentences():
    base_dir = BASE_DIR

    with open(base_dir / 'data/raw_data/dataset.json', 'r', encoding='utf-8') as f:
        data = json.load(f)

    all_samples = []
    for case in data:
        samples = process_case_sections(case)
        all_samples.extend(samples)

    label_counts = Counter(sample['label'] for sample in all_samples)
    print("\nInitial category distribution:")
    for label, count in label_counts.items():
        print(f"{label}: {count}")

    balanced_samples = balance_samples(all_samples)

    label_counts = Counter(sample['label'] for sample in balanced_samples)
    print("\nBalanced category distribution:")
    for label, count in label_counts.items():
        print(f"{label}: {count}")

    output_path = base_dir / "data/processed_data"
    output_path.mkdir(parents=True, exist_ok=True)

    with open(output_path / "sentence_samples.json", 'w', encoding='utf-8') as f:
        json.dump(balanced_samples, f, indent=2, ensure_ascii=False)

    print(f"\nProcessed and balanced {len(balanced_samples)} sentences across {len(data)} cases")
```

## Preprocessing (with SpaCy)

```python
def split_into_sentences_spacy(text: str, min_words=6) -> list[str]:
    if not text.strip():
        return []

    # Normalize quotes, dashes, citations
    text = normalize_text(text)
    doc = nlp(text)

    sentences = []
    for sent in doc.sents:
        cleaned = clean_serial_numbers(sent.text).strip()
        if len(cleaned.split()) >= min_words and not is_boilerplate(cleaned):
            sentences.append(cleaned)
    return sentences
```

```python
def process_case_sections_spacy(case: dict) -> list[dict]:
    sections = extract_sections(case)
    samples = []

    section_types = {
        "facts": "FACT",
        "issues": "ISSUE",
        "precedent_analysis": "PRECEDENT",
        "court_reasoning": "REASONING",
        "conclusion": "CONCLUSION"
    }

    for section_name, label in section_types.items():
        text = sections[section_name]
        text = normalize_text(text)
        sentences = split_into_sentences_spacy(text)

        for idx, sentence in enumerate(sentences):
            if sentence.strip() and not is_boilerplate(sentence):
                context = get_sentence_context(sentences, idx)
                sample = {
                    "text": sentence.strip(),
                    "label": label,
                    "metadata": {
                        "case_id": sections["id"],
                        "case_name": sections["case_name"],
                        "section": section_name,
                        "context": context,
                        "position": f"{idx + 1}/{len(sentences)}",
                        "has_citation": bool(re.search(LEGAL_CITATIONS, sentence))
                    }
                }
                samples.append(sample)

    return samples

def preprocess_sentences_spacy():
    base_dir = BASE_DIR

    with open(base_dir / 'data/raw_data/dataset.json', 'r', encoding='utf-8') as f:
        data = json.load(f)

    all_samples = []
    for case in data:
        samples = process_case_sections_spacy(case)
        all_samples.extend(samples)

    label_counts = Counter(sample['label'] for sample in all_samples)
    print("\nInitial category distribution:")
    for label, count in label_counts.items():
        print(f"{label}: {count}")

    balanced_samples = balance_samples(all_samples)

    label_counts = Counter(sample['label'] for sample in balanced_samples)
    print("\nBalanced category distribution:")
    for label, count in label_counts.items():
        print(f"{label}: {count}")

    output_path = base_dir / "data/processed_data"
    output_path.mkdir(parents=True, exist_ok=True)

    with open(output_path / "sentence_samples.json", 'w', encoding='utf-8') as f:
        json.dump(balanced_samples, f, indent=2, ensure_ascii=False)

    print(f"\nProcessed and balanced {len(balanced_samples)} sentences across {len(data)} cases")
```

### Training Model Functions

```python
def compute_metrics(eval_pred):
    predictions, labels = eval_pred
    predictions = np.argmax(predictions, axis=1)
    return {
        'accuracy': (predictions == labels).mean(),
    }

def train_model(model_name: str, save_path: Path):
    texts, labels, label2id, id2label = load_data()

    train_texts, val_texts, train_labels, val_labels = train_test_split(
        texts, labels, test_size=0.2, random_state=42
    )

    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModelForSequenceClassification.from_pretrained(
        model_name,
        num_labels=len(label2id),
        id2label=id2label,
        label2id=label2id,
        from_tf=True
    )

    train_dataset = LegalSentenceDataset(train_texts, train_labels, tokenizer, max_length=256)
    val_dataset = LegalSentenceDataset(val_texts, val_labels, tokenizer, max_length=256)

    training_args = TrainingArguments(
        output_dir=str(save_path / "temp"),  # temporary training output
        eval_strategy="epoch",
        learning_rate=1e-5,
        per_device_train_batch_size=16,
        per_device_eval_batch_size=16,
        num_train_epochs=3,
        weight_decay=0.01,
        logging_steps=100,
        save_strategy="no", # no checkpoints because google drive has limited space
        load_best_model_at_end=False
    )

    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=train_dataset,
        eval_dataset=val_dataset,
        compute_metrics=compute_metrics,
    )

    trainer.train()

    save_path.mkdir(parents=True, exist_ok=True)
    model.config.save_pretrained(save_path)
    tokenizer.save_pretrained(save_path)
    trainer.save_model(save_path)

    return trainer, val_dataset, val_labels, id2label
```

### Evaluating performance

```python
def evaluate_model(trainer, val_dataset, val_labels, id2label):
    predictions = trainer.predict(val_dataset)
    preds = np.argmax(predictions.predictions, axis=1)

    print("\nClassification Report:\n")
    print(classification_report(
        val_labels,
        preds,
        target_names=[id2label[i] for i in range(len(id2label))]
    ))
```

# Classification

Below is the class definition for the classifier that we shall use for annotation purposes.

```python
# RUN IF NOT ALREADY MOUNTED
# from google.colab import drive
# drive.mount('/content/drive')
```

```python
class LegalSentenceClassifier:
    def __init__(self, model_path: Path):
        self.device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
        self.model_path = model_path

        # load tokenizer and model
        self.tokenizer = AutoTokenizer.from_pretrained(model_path)
        self.model = AutoModelForSequenceClassification.from_pretrained(model_path).to(self.device)
        self.model.eval()

        self.id2label = self.model.config.id2label

    def classify(self, text: str) -> tuple[str, float]:
        inputs = self.tokenizer(text, return_tensors="pt", truncation=True, max_length=256).to(self.device)

        with torch.no_grad():
            outputs = self.model(**inputs)
            probs = torch.softmax(outputs.logits, dim=1)
            prediction = torch.argmax(probs, dim=1)
            confidence = torch.max(probs).item()

        predicted_label = self.id2label[prediction.item()]
        return predicted_label, confidence
```

# Annotation

## Annotator Class Definition

```python
class PDFAnnotator:
    def __init__(self, model_path=None):
        self.classifier = LegalSentenceClassifier(model_path)
        self.label_colors = {
            'FACT': red,
            'ISSUE': blue,
            'PRECEDENT': green,
            'REASONING': orange,
            'CONCLUSION': purple
        }

        self.min_confidence = 0.65 # MINIMUM CONFIDENCE REQUIRED TO HIGHLIGHT

    def _extract_text_with_positions(self, pdf_path):
        pages_data = []
        with pdfplumber.open(pdf_path) as pdf:
            for page in pdf.pages:
                pages_data.append({
                    'text': page.extract_text(),
                    'words': page.extract_words(),
                    'height': float(page.height),
                    'width': float(page.width)
                })
        return pages_data

    def _find_sentence_bounds(self, sentence: str, words: list) -> tuple:
        clean_sentence = ' '.join(sentence.split())
        found_words = []
        current_match = []

        for word in words:
            current_match.append(word)
            current_text = ' '.join(w['text'] for w in current_match)

            if clean_sentence in current_text:
                found_words = current_match
                break

            if len(current_text) > len(clean_sentence) + 10:
                current_match.pop(0)

        if found_words:
            return (
                min(float(w['x0']) for w in found_words),
                min(float(w['top']) for w in found_words),
                max(float(w['x1']) for w in found_words),
                max(float(w['bottom']) for w in found_words)
            )
        return None

    def _rgb_to_color(self, color):
        return (color.red, color.green, color.blue)

    def annotate_pdf(self, input_path, output_path):
        print("Extracting text from PDF...")
        pages_data = self._extract_text_with_positions(input_path)
        pdf = fitz.open(input_path)

        print("Processing and classifying sentences...")
        for page_num, page_data in enumerate(pages_data):
            page = pdf[page_num]
            sentences = sentence_splitter(page_data['text'])

            for sentence in sentences:
                try:
                    bounds = self._find_sentence_bounds(sentence, page_data['words'])
                    if not bounds or len(sentence.split()) < 8:
                        continue

                    x0, y0, x1, y1 = bounds
                    normalized = normalize_text(sentence)
                    label, confidence = self.classifier.classify(normalized)

                    if confidence > self.min_confidence:
                        rect = fitz.Rect(x0, y0, x1, y1)
                        color = self._rgb_to_color(self.label_colors[label])
                        annot = page.add_highlight_annot(rect)
                        annot.set_colors(stroke=color)
                        annot.set_opacity(0.3)
                        annot.set_info(title="Classification", content=f"{label} ({confidence:.1%})")
                        annot.update()

                except Exception as e:
                    print(f"Error processing sentence: {e}")

        pdf.save(output_path)
        pdf.close()
        print(f"Annotated PDF saved to: {output_path}")
```

## Directory Setup

```python
def setup_directories():
    base_dir = BASE_DIR
    pdf_dir = base_dir / "pdfs"
    test_pdf_dir = pdf_dir / "test_pdfs"
    annotated_dir = pdf_dir / "annotated_pdfs"

    pdf_dir.mkdir(exist_ok=True)
    test_pdf_dir.mkdir(exist_ok=True)
    annotated_dir.mkdir(exist_ok=True)

    return pdf_dir, test_pdf_dir, annotated_dir
```

## Process PDFs

```python
models_dir = BASE_DIR / "models"
available_models = [f.name for f in models_dir.iterdir() if f.is_dir()]
if not available_models:
    print("No models found in 'models/' directory.")
else:
    print("Available models:")
    for i, model in enumerate(available_models, 1):
        print(f"{i}. {model}")

    # user picks a model by name
    selected_model = input("\nEnter the model folder name to use: ").strip()
    model_path = models_dir / selected_model

    if not model_path.exists():
        print(f"\nModel path '{model_path}' does not exist.")
    else:
        pdf_dir, test_pdf_dir, annotated_dir = setup_directories()
        annotator = PDFAnnotator(model_path=model_path)

        pdf_files = [f for f in test_pdf_dir.iterdir() if f.suffix.lower() == ".pdf"]


        if not pdf_files:
            print(f"No PDF files found in {test_pdf_dir}")
        else:
            to_process = []
            for input_path in pdf_files:
                output_path = annotated_dir / f"{input_path.stem}_annotated.pdf"
                if not output_path.exists():
                    to_process.append(input_path)
                else:
                    print(f"Skipping {input_path.name} (already annotated)")

            if not to_process:
                print("No new PDFs to process.")
            else:
                print(f"\nFound {len(to_process)} PDFs to process.")
                for input_path in to_process:
                    print(f"\nProcessing: {input_path.name}")
                    output_path = annotated_dir / f"{input_path.stem}_annotated.pdf"
                    try:
                        annotator.annotate_pdf(str(input_path), str(output_path))
                    except Exception as e:
                        print(f"Error processing {input_path.name}: {e}")
                print("\nAll files processed.")
                print(f"Annotated PDFs saved to: {annotated_dir}")
```
